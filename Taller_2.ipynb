```python
class Estudiante:
    def __init__(self, nombre, apellido, cedula, email, semestre, materias):
        self.nombre = nombre
        self.apellido = apellido
        self.cedula = cedula
        self.email = email
        self.semestre = semestre
        self.materias = materias

    # Nombre
    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, value):
        self._nombre = value

    # Apellido
    @property
    def apellido(self):
        return self._apellido

    @apellido.setter
    def apellido(self, value):
        self._apellido = value

    # Cédula (VALIDACIÓN)
    @property
    def cedula(self):
        return self._cedula

    @cedula.setter
    def cedula(self, value):
        if value.isdigit() and len(value) == 10:
            self._cedula = value
        else:
            print("Cédula inválida")

    # Email
    @property
    def email(self):
        return self._email

    @email.setter
    def email(self, value):
        self._email = value

    # Semestre (VALIDACIÓN)
    @property
    def semestre(self):
        return self._semestre

    @semestre.setter
    def semestre(self, value):
        if value >= 1 and value <= 9:
            self._semestre = value
        else:
            print("Semestre inválido")

    # Materias
    @property
    def materias(self):
        return self._materias

    @materias.setter
    def materias(self, value):
        self._materias = value

    def __str__(self):
        return f"""Nombre: {self.nombre}
Apellido: {self.apellido}
Cédula: {self.cedula}
Email: {self.email}
Semestre: {self.semestre}
Materias: {self.materias}"""


if __name__ == "__main__":
    vali = Estudiante(
        "Akyra",
        "Chalen",
        "0952850105",
        "shuskac@gmail.com",
        3,
        ["POO"]
    )

    print(vali)
```
